```python
import pandas as pd
```

```python
# Load data
zhvi_allhomes = pd.read_csv('../data/raw/Metro_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
zhvi_singlefamily = pd.read_csv('../data/raw/Metro_zhvi_uc_sfr_tier_0.33_0.67_sm_sa_month.csv')
zhvi_condo = pd.read_csv('../data/raw/Metro_zhvi_uc_condo_tier_0.33_0.67_sm_sa_month.csv')
sales_count = pd.read_csv('../data/raw/Metro_sales_count_raw_uc_sfr_month.csv')
inventory_count = pd.read_csv('../data/raw/Metro_invt_fs_uc_sfr_month.csv')
```

```python
# List of columns to keep
id_vars = ['RegionID', 'RegionName', 'RegionType', 'StateName', 'SizeRank']

# MELT each dataframe (wide -> long)
zhvi_allhomes_melted = zhvi_allhomes.melt(
    id_vars=id_vars,
    var_name='Date',
    value_name='ZHVI_AllHomes'
)

zhvi_singlefamily_melted = zhvi_singlefamily.melt(
    id_vars=id_vars,
    var_name='Date',
    value_name='ZHVI_SingleFamily'
)

zhvi_condo_melted = zhvi_condo.melt(
    id_vars=id_vars,
    var_name='Date',
    value_name='ZHVI_Condo'
)

sales_count_melted = sales_count.melt(
    id_vars=id_vars,
    var_name='Date',
    value_name='SalesCount_SFR'
)

inventory_count_melted = inventory_count.melt(
    id_vars=id_vars,
    var_name='Date',
    value_name='Inventory_SFR'
)
```

```python
# Merge all melted data correctly by RegionName and Date
merged_df = zhvi_allhomes_melted.merge(
    zhvi_singlefamily_melted[['RegionName', 'Date', 'ZHVI_SingleFamily']],
    on=['RegionName', 'Date'],
    how='left'
).merge(
    zhvi_condo_melted[['RegionName', 'Date', 'ZHVI_Condo']],
    on=['RegionName', 'Date'],
    how='left'
).merge(
    sales_count_melted[['RegionName', 'Date', 'SalesCount_SFR']],
    on=['RegionName', 'Date'],
    how='left'
).merge(
    inventory_count_melted[['RegionName', 'Date', 'Inventory_SFR']],
    on=['RegionName', 'Date'],
    how='left'
)

print(merged_df.head())
```

```python
merged_df.columns
```

```python
# Convert 'Date' column to datetime type if not already
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Define the key columns to check for missing values
cols_to_check = ['ZHVI_AllHomes', 'ZHVI_SingleFamily', 'ZHVI_Condo', 'SalesCount_SFR', 'Inventory_SFR']

# Drop rows with any missing values in key columns, group by date, and count valid rows
valid_dates = merged_df.dropna(subset=cols_to_check).groupby('Date').size()

# Find the earliest date where all columns have non-missing values
first_valid_date = valid_dates.index.min()

print(f"First valid date with full data across all metrics: {first_valid_date.date()}")

# Store original number of rows
original_row_count = merged_df.shape[0]

# Filter the dataframe to include only rows from the first valid date and onward
merged_df_filtered = merged_df[merged_df['Date'] >= first_valid_date].copy()

# Store new number of rows after filtering
filtered_row_count = merged_df_filtered.shape[0]

# Print the difference
print(f"Rows before filtering:   {original_row_count}")
print(f"Rows after filtering:    {filtered_row_count}")
print(f"Rows removed:            {original_row_count - filtered_row_count}")
```

```python
# Find one top-ranked metro area per state
top_regions = (
    merged_df_filtered[merged_df_filtered['RegionName'] != 'United States']
    .sort_values(['StateName', 'SizeRank'])
    .groupby('StateName', as_index=False)
    .first()
)[['RegionName', 'StateName']]

# Keep rows matching top regions OR the United States row
mask = merged_df_filtered[['RegionName', 'StateName']].isin(top_regions.to_dict('list')).all(axis=1)
us_rows = merged_df_filtered['RegionName'] == 'United States'

# Combine filtered metro areas + national data
merged_df_top_plus_us = merged_df_filtered[mask | us_rows].copy()


original_row_count = merged_df_filtered.shape[0]
filtered_row_count = merged_df_top_plus_us.shape[0]

print(f"Rows before sorting:   {original_row_count}")
print(f"Rows after sorting:    {filtered_row_count}")
print(f"Rows removed:          {original_row_count - filtered_row_count}")

# print(merged_df_top_cities[['StateName', 'RegionName']].drop_duplicates().sort_values('StateName'))
```

```python
# Save merged DataFrame
merged_df_top_plus_us.to_csv('../data/processed/01_housing_data_merged.csv', index=False)

print("File successfully saved to 'data/processed/01_housing_data_merged.csv'")
```
